## Завдання 1: Розрахунок днів між датами

Створіть функцію `get_days_from_today(date)`, яка розраховує кількість днів між заданою датою і поточною датою.

**Вимоги до завдання:**
1. Функція приймає один параметр: `date` — рядок, що представляє дату у форматі 'РРРР-ММ-ДД' (наприклад, '2020-10-09').
2. Функція повертає ціле число, яке вказує на кількість днів від заданої дати до поточної.
   - Якщо задана дата пізніша за поточну, результат має бути від'ємним.
3. У розрахунках необхідно враховувати лише дні, ігноруючи час (години, хвилини, секунди).
4. Для роботи з датами слід використовувати модуль `datetime` Python.

**Приклад:**
Якщо сьогодні 5 травня 2021 року, виклик `get_days_from_today("2021-10-09")` повинен повернути 157, оскільки 9 жовтня 2021 року є на 157 днів пізніше від 5 травня 2021 року.


### Рекомендації для виконання (Завдання 1)

1. Імпортуйте модуль `datetime`.
2. Перетворіть рядок дати у форматі 'РРРР-ММ-ДД' у об'єкт `datetime`.
3. Отримайте поточну дату, використовуючи `datetime.today()`.
4. Розрахуйте різницю між поточною датою та заданою датою.
5. Поверніть різницю у днях як ціле число.


````python
# Ваше рішення для Завдання 1 тут
import datetime

def get_days_from_today(date):
    date_obdg = datetime.datetime.strptime(date, "%Y-%m-%d").date()
    date_now = datetime.date.today()
    delta = date_now - date_obdg
    return delta.days
````

### Рекомендації для виконання (Завдання 2)

1. Переконайтеся, що вхідні параметри відповідають заданим обмеженням (min >= 1, max <= 1000, min < quantity < max).
2. Використовуйте модуль `random` для генерації випадкових чисел.
3. Використовуйте множину (`set`) або інший механізм для забезпечення унікальності чисел.
4. Пам'ятайте, що функція має повертати **відсортований** список унікальних чисел.


````python
# Ваше рішення для Завдання 2 тут
import random

def get_numbers_ticket(min, max, quantity):
    if min < 1 or max > 1000 or max < quantity < min:
      return []

    random_numbers = random.sample(range(min, max+1), quantity)
    return sorted(random_numbers)
````

````python
lottery_numbers = get_numbers_ticket(1, 1001, 6)
print("Ваші лотерейні числа:", lottery_numbers)
````

## Завдання 3: Нормалізація телефонних номерів

Ваш сервіс розсилок може ефективно відправляти SMS лише тоді, коли номери телефонів представлені у коректному форматі. Вам необхідна функція, яка автоматично нормалізує номери телефонів, видаляючи зайві символи та додаючи код країни.

Розробіть функцію `normalize_phone(phone_number)`, що нормалізує телефонні номери до стандартного формату, залишаючи тільки цифри та символ '+' на початку.

**Вимоги до завдання:**
1. Параметр функції `phone_number` — це рядок з телефонним номером у різноманітних форматах.
2. Функція видаляє всі символи, крім цифр та символу '+'.
3. Якщо міжнародний код відсутній, функція додає код '+38' (для України). Це враховує випадки, коли номер починається з '380' (додається лише '+') та коли номер починається без коду (додається '+38').
4. Функція повертає нормалізований телефонний номер у вигляді рядка.

**Приклад використання:**
```python
raw_numbers = [
    "067\t123 4567",
    "(095) 234-5678\n",
    "+380 44 123 4567",
    "380501234567",
    "    +38(050)123-32-34",
    "0503451234",
    "(050)8889900",
    "38050-111-22-22",
    "38050 111 22 11   ",
]

sanitized_numbers = [normalize_phone(num) for num in raw_numbers]
print("Нормалізовані номери:", sanitized_numbers)
# Очікується: ['+380671234567', '+380952345678', '+380441234567', ...]
```

### Рекомендації для виконання (Завдання 3)

1. Використовуйте модуль `re` для регулярних виразів для видалення непотрібних символів.
2. Перевірте, чи номер починається з '+', і виправте префікс згідно з вказівками.
3. Видаліть всі символи, крім цифр та '+', з номера телефону.
4. Не забувайте повертати нормалізований номер телефону з функції.


````python
# Ваше рішення для Завдання 3 тут
import re

def normalize_phone(phone_number):
    clear_number = re.sub(r'[^\d]', "", phone_number)
    if clear_number.startswith("380"):
        return "+" + clear_number
    else:
        return "+38" + clear_number
````

````python
raw_numbers = [
    "067\t123 4567",
    "(095) 234-5678\n",
    "+380 44 123 4567",
    "380501234567",
    "    +38(050)123-32-34",
    "0503451234",
    "(050)8889900",
    "38050-111-22-22",
    "38050 111 22 11   ",
]

sanitized_numbers = [normalize_phone(num) for num in raw_numbers]
print("Нормалізовані номери:", sanitized_numbers)
````

## Завдання 4: Список привітань (Дні народження)

Вам потрібно створити функцію `get_upcoming_birthdays(users)`, яка допоможе визначати, кого з колег потрібно привітати. Функція повинна повернути список всіх, у кого день народження вперед на 7 днів включаючи поточний день.

**Вимоги до завдання:**
1. Параметр `users` — це список словників, де кожен словник містить ключі `name` (ім'я користувача, рядок) та `birthday` (день народження, рядок у форматі 'рік.місяць.дата').
2. Функція визначає, кого потрібно привітати протягом наступних 7 днів (включаючи поточний день).
3. **Важливо:** Якщо день народження припадає на вихідний, дата привітання переноситься на наступний понеділок.
4. Функція повертає список словників, де кожен словник містить інформацію про користувача (`name`) та дату привітання (`congratulation_date`, рядок 'РРРР.ММ.ДД').

**Приклад вхідних даних:**
```python
users = [
    {"name": "John Doe", "birthday": "1985.01.23"},
    {"name": "Jane Smith", "birthday": "1990.01.27"}
]
[
    {'name': 'John Doe', 'congratulation_date': '2024.01.23'},
    {'name': 'Jane Smith', 'congratulation_date': '2024.01.29'}
]
```


### Рекомендації для виконання (Завдання 4)

1. Приймайте список `users`, де кожен елемент містить `birthday` у форматі рядка ("YYYY.MM.DD").
2. Перетворіть дати народження з рядків у об'єкти `datetime` (використовуйте `datetime.strptime` та `.date()`).
3. Визначте поточну дату системи за допомогою `datetime.today().date()`.
4. Пройдіться по списку `users`. Для кожного користувача:
    - Перевірте, чи вже минув день народження в цьому році. Якщо так — розглядайте дату на наступний рік.
    - Визначте різницю між днем народження та поточним днем.
    - Перевірте, чи входить ця дата у діапазон наступних 7 днів.
5. Перевірте, чи день народження припадає на вихідний (субота або неділя). Якщо так — перенесіть дату привітання на понеділок.
6. Створіть структуру даних (словник) з ім'ям та датою привітання і додайте її до списку результатів.


````python
from datetime import datetime, date, timedelta
from typing import List, Dict

def fint_next_birthday(birthday_date:str, date_now:date)->date:
  """
  Розраховує наступну дату дня народження для заданого користувача.

  Аргументи:
    birthday_date (str): Рядок дати народження у форматі 'РРРР.ММ.ДД'.
    date_now (date): Поточна дата.

  Повертає:
    date: Наступна дата дня народження, скоригована для поточного року.
  """
  birthday_date = datetime.strptime(birthday_date, "%Y.%m.%d").date()
  birthday_this_year = birthday_date.replace(year=date_now.year)
  if birthday_this_year < date_now:
    birthday_this_year = birthday_date.replace(year=date_now.year+1)
  return birthday_this_year


def adjust_for_weekend(birthday:date)->date:
  """
  Коригує дату дня народження, якщо вона припадає на вихідний, переносячи її на наступний понеділок.

  Аргументи:
    birthday (date): Оригінальна дата дня народження.

  Повертає:
    date: Скоригована дата дня народження (понеділок, якщо це був вихідний, інакше оригінальна дата).
  """
  if birthday.weekday() == 5: # Saturday
    return birthday+timedelta(days=2)
  elif birthday.weekday() == 6: # Sunday
    return birthday+timedelta(days=1)
  return birthday

def get_upcoming_birthdays(users:List[Dict[str, str]])->List[Dict[str, str]]:
  """
  Визначає майбутні дні народження протягом наступних 7 днів, включаючи сьогодні.
  Коригує дати святкування на вихідні, переносячи їх на наступний понеділок.

  Аргументи:
    users (List[Dict[str, str]]): Список словників, кожен з яких містить 'name' (ім'я) та 'birthday' (день народження) у форматі 'РРРР.ММ.ДД'.

  Повертає:
    List[Dict[str, str]]: Список словників, кожен з яких містить 'name' (ім'я) та 'congratulation_date' (дата привітання) у форматі 'РРРР.ММ.ДД'
                          для користувачів з майбутніми днями народження.
  """
  upcoming_birthdays = []
  data_now = date.today()
  for user in users:
    birthday_date = fint_next_birthday(user["birthday"], data_now)
    delta = (birthday_date - data_now).days
    if 0 <= delta <= 7:
      conk_date =  adjust_for_weekend(birthday_date)
      result = {
          "name": user["name"],
          "congratulation_date": conk_date.strftime("%Y.%m.%d")
      }
      upcoming_birthdays.append(result)
  return upcoming_birthdays
````

````python
users = [
    {"name": "John Doe", "birthday": "1985.02.13"},
    {"name": "Jane Smith", "birthday": "1990.02.15"}
]
get_upcoming_birthdays(users)
````
